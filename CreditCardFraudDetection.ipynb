```python
#number of missing values
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fradulent transaction
#o represents legit transaction and 1 represents fradulent transaction
credit_card_data['Class'].value_counts()
```

This data set is unbalanced
0---->Normal Transaction
1---->Fradulent Transaction

```python
#seperating data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical data
legit.Amount.describe()
#below given are not percentages they are percentile
```

```python
fraud.Amount.describe()
```

```python
#compare the value of both the transaction
credit_card_data.groupby('Class').mean()
```

Under-Sampling
Build a sample dataset containing similar distribution of normal transaction and the fradulent transaction

```python
legit_sample=legit.sample(n=38)
```

Concatenate two dataframes

```python
new_dataset=pd.concat((legit_sample,fraud),axis=0)
#axis =0  means the data is added row wise
#axis =1 means the data is added column wise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
#this step is very important to ensure that do we got good sample or bad sample
#if the mean values are very diffrent from the mean from the original sample the data is unbalanced else the dataset is balanced
```

Splitting the data into features and targets

Target are 0 and 1
Features are v1 v2 v3 and so on

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training data and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_test.shape,X_train.shape)
```

Model Training  Logistic Regression

```python
model=LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training Data : ',training_data_accuracy)
```

```python
#Accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on Test Data : ',test_data_accuracy)
```

IF THE DIFFRENCE BETWEEN THE TRAINING DATASET AND TESTING DATASET IS LARGE THAN THE MODEL IS NOT GOOD BUT IF THE DIFFRENCE IS SMALL THAN THE MODEL IS GOOD
